# Finding Best Random State

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
maxaccu=0
maxRs=0
for i in range(0,200):
    fea_res1_train,fea_res1_test,tar_res1_train,tar_res1_test=train_test_split(fea_res1,tar_res1,test_size=.20,random_state=i)
    dt.fit(fea_res1_train,tar_res1_train)
    pre_train=dt.predict(fea_res1_test)
    acc=accuracy_score(tar_res1_test,pre_train)
    print('accuracy',acc,'random_state',i)
    
    if acc>maxaccu:
        maxaccu=acc
        maxRs=i
        print('maximum accuracy',maxaccu,'maximum random state',maxRs)
```

```python
print('best accuracy score',maxaccu ,'on random state',maxRs)
```
